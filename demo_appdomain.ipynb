```python
from appdomain import *
from appdomain.ensemble import ApplicabilityDomain
import numpy as np
import matplotlib.pyplot as plt
```

```python
knn = kNNApplicabilityDomain()
ocsvm = OCSVMApplicabilityDomain()
lof = LOFApplicabilityDomain()
irf = iForestApplicabilityDomain()

cad = ApplicabilityDomain()
```

```python
# generation of samples
n_samples_1 = 500
mean_1 = [0, 0]
cov_1 = [[0.3, 0.28], [0.28, 0.3]]

n_samples_2 = 50
mean_2 = [-2, 2]
cov_2 = [[0.5, 0], [0, 0.5]]

np.random.seed(11)
x_1 = np.random.multivariate_normal(mean_1, cov_1, n_samples_1)
x_2 = np.random.multivariate_normal(mean_2, cov_2, n_samples_2)
np.random.seed()
x = np.r_[x_1, x_2]

plt.rcParams['font.size'] = 18 
plt.scatter(x[:n_samples_1, 0], x[:n_samples_1, 1])
plt.scatter(x[n_samples_1:, 0], x[n_samples_1:, 1])
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim([-4.5, 2.5])
plt.ylim([-2.5, 4.5])
plt.show()
```

```python
train = x[:400]
test = x[400:]
```

```python
def fit_and_show(model):
    model.fit(train)
    print(model.predict(test))

    plt.scatter(test[:, 0], test[:,1], c=model.predict(test), cmap='tab10')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.xlim([-4.5, 2.5])
    plt.ylim([-2.5, 4.5])
    plt.show()
```

```python
fit_and_show(knn)
```

```python
fit_and_show(ocsvm)
```

```python
fit_and_show(lof)
```

```python
fit_and_show(irf)
```

```python
fit_and_show(cad)
```
